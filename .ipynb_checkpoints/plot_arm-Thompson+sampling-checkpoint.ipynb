```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
#########################################################################
pylab.plot(x,w34,'b')
plt.axvline(x=0.5586, ymin=0.04, ymax=0.56, color='r', linestyle='--')
plt.xlabel('Arm (weight)')
plt.ylabel('Beta distribution')
plt.text(1, 2, r'$w_{34}$')
plt.text(0.42, 0.2, r'$Random \ Sample$')

#plt.axis([40, 160, 0, 0.03])
plt.grid(True)
pylab.show()
```

```python
# Draw the propobilty distribaution of beta using scipy
import numpy as np
import pylab
from scipy.stats import beta
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from matplotlib import pyplot as plt

x = np.linspace(-0.02,1.02,1000)

w21 = beta.pdf(x,  3,  2)    


w34 = beta.pdf(x,  2,  1)

```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
#########################################################################
pylab.plot(x,w21,'b')
plt.axvline(x=0.6812, ymin=0.04, ymax=0.95, color='r', linestyle='--')
plt.xlabel('Arm (weight)')
plt.ylabel('Beta distribution')
plt.text(0.9, 1, r'$w_{21}$')
plt.text(0.58, 0.5, r'$Random \ Sample$')

#plt.axis([40, 160, 0, 0.03])
plt.grid(True)
pylab.show()
```
