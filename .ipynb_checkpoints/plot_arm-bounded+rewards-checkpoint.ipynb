```python
import numpy as np
from scipy.stats import norm
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from matplotlib import pyplot as plt

x = np.linspace(-1,20,1000)
y1 = norm.pdf(x, loc=4.0, scale=1.5)    # for example
y2 = norm.pdf(x, loc=12, scale=1.0)    # for example
y3 = norm.pdf(x, loc=5, scale=2.0)    # for example
y4 = norm.pdf(x, loc=7.0, scale=1.3)    # for example
y5 = norm.pdf(x, loc=13.0, scale=1.0)    # for example
y6 = norm.pdf(x, loc=10.0, scale=1.0)    # for example
y7 = norm.pdf(x, loc=8.0, scale=0.5)    # for example
y8 = norm.pdf(x, loc=9.0, scale=2.0)    # for example
y9 = norm.pdf(x, loc=10.0, scale=1.5)    # for example
y10 = norm.pdf(x, loc=6.0, scale=1.8)    # for example

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size


plt.plot(x,y1, x,y2, x,y3, x,y4, x,y5,x,y6,x,y7,x,y8,x,y9,x,y10)
plt.show()
```
