# Prepare the data for classification

```python
dataset = df.values
```

```python
from sklearn import preprocessing
# for training data
X = dataset.astype(float)
features = preprocessing.scale(X)
target = y
```

```python
X.shape
```

# Feature Selection

```python
from sklearn.cross_validation import train_test_split
# Extracting 20% validation data
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0, random_state=17)
```

## Preproccing

```python
print('Number of training examples',len(X_train))
print('Number of validation examples',len(X_test))
```

```python
from sklearn import preprocessing
# normalize the data attributes
X_train = preprocessing.normalize(X_train)
#X_test = preprocessing.normalize(X_test)
# standardize the data attributes
X_train = preprocessing.scale(X_train)
#X_test = preprocessing.scale(X_test)
```

# Neural Networks

```python
import seaborn as sns
from sklearn.cross_validation import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.regularizers import l2
from keras.utils import np_utils
labelsTrain = np_utils.to_categorical(y_train)
model = Sequential()
model.add(Dense(16,
                input_shape=(2,), 
                activation="relu"))
model.add(Dense(2, activation="softmax"))
model.compile(loss='categorical_crossentropy',
              metrics=['accuracy'],
              optimizer='adam')
start_time = time.time()
model.fit(X_train, labelsTrain, verbose=0, batch_size=1, nb_epoch=100)
print("The time for training NN is  %s seconds " % (time.time() - start_time))
loss, accuracy = model.evaluate(X_train, labelsTrain, batch_size=1, verbose=0)
print("Test fraction correct (NN-loss) = {:.2f}".format(loss))
print("Test fraction correct (NN-Accuracy) = {:.2f}".format(accuracy))
```

```python
modelBuckup = model
```

# Direct Method Pruning

```python
N_arry = np.zeros((10,16))
N_arryBinar = np.zeros((10,16))
```

```python
Threshold = 0
constant=1
r=0
for k in range(0, 1000, 100):
    sample_x = X_train[k:k+100]
    sample_y = labelsTrain[k:k+100]
    for i in range(0, 16): 
        for b in range(0,100):
            loss, accuracy = modelBuckup.evaluate(sample_x[b:b+1], sample_y[b:b+1], 
                                                 batch_size=1, verbose=0)
        # Prune the neuron in the layer
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(sample_x[b:b+1], sample_y[b:b+1], 
                                                         batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            print('delta = ', delta)

            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
        
            #print(r)
            val = i+1
            N_arry[r][i] = (val-1)/val * N_arry[r][i] + 1/val * reward
            #N_arry[r][i] = N_arry[r][i] + reward
            if reward>0:
                R=1
            else:
                R=0
            N_arryBinar[r][i] = (val-1)/val * N_arryBinar[r][i] + 1/val * R
            #N_arryBinar[r][i] = N_arryBinar[r][i] + R
                 
            modelBuckup.set_weights(All_weights)
            
                #print(delta)
    
    r=r+1
    #print(r)
```

# Epsilon Greedy

```python
x = np.arange(0,16)
def ploting_function(i, R, P, chosen):    
    plt.scatter(x, R)
    plt.scatter(chosen, R[chosen], c='r', label='Rnd = ' + str(P))
    #plt.scatter(chosen, R[chosen], c='r')
    plt.xlabel('Arms (t = '+ str(i)+')' + ' and ' + 'epsilon = 0.5')
 
    plt.legend(loc='upper left')
    
    plt.ylabel('The mean')
    plt.xticks(x,column)
    plt.savefig("/Users/salemameen/Desktop/directvsMAB/e_greedy/Fig"+str(i)+".png")
    plt.clf()
    #plt.show()
```

## Pruning the Weights

```python
Avg_Accumaltive_R_EGN = np.zeros(16)
Count_EGN = np.zeros(16)
import random
epsilon=0.5
for i in range(0, 16): 
            loss, accuracy = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            Avg_Accumaltive_R_EGN[i] = reward
            modelBuckup.set_weights(All_weights)
            Rnd = 0
            ploting_function(i+1, Avg_Accumaltive_R_EGN, Rnd, i)


for j in range(17, 1001):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            Rnd = random.uniform(0, 1)
            if (epsilon>Rnd):
                i = np.argmax(Avg_Accumaltive_R_EGN)
                
            else:
                i=random.randint(0,15) 
            Count_EGN[i] = Count_EGN[i] + 1
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            val = Count_EGN[i]
            Avg_Accumaltive_R_EGN[i] = (val-1)/val * Avg_Accumaltive_R_EGN[i] + 1/val * reward
            modelBuckup.set_weights(All_weights)
            
            ploting_function(j, Avg_Accumaltive_R_EGN, Rnd, i)

df5 = pandas.DataFrame(Avg_Accumaltive_R_EGN)
df5['Weights']=column
df5.set_index('Weights')
df5.reset_index('Weights')
df5.head(16)
```

## Pruning the Weights

```python
Avg_Accumaltive_R_SGN = np.zeros(16)
Count_SGN = np.zeros(16)

import random
tau=4
for i in range(0, 16): 
            loss, accuracy = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            Avg_Accumaltive_R_SGN[i] = reward
            modelBuckup.set_weights(All_weights)
            ploting_function(i+1, Avg_Accumaltive_R_SGN, 0.05, i)


for j in range(17, 1001):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            prob = np.exp(Avg_Accumaltive_R_SGN/tau)
            sum = prob.sum()
            prb = prob/sum
            i = np.random.choice(numpy.arange(0, 16), p=prb.reshape(16))
            
            Count_SGN[i] = Count_SGN[i] + 1
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            val = Count_EGN[i]
            Avg_Accumaltive_R_SGN[i] = (val-1)/val * Avg_Accumaltive_R_SGN[i] + 1/val * reward
            modelBuckup.set_weights(All_weights)
            
            ploting_function(j, Avg_Accumaltive_R_SGN, prb, i)

df5_S = pandas.DataFrame(Avg_Accumaltive_R_SGN)
df5_S['Weights']=column
df5_S.set_index('Weights')
df5_S.reset_index('Weights')
df5_S.head(16)
```

# Hedge Algorithm

```python
x = np.arange(0,16)
def ploting_function(i, R, chosen):    
    plt.scatter(x, R)
    plt.scatter(chosen, R[chosen], c='r')
    #plt.scatter(chosen, R[chosen], c='r')
    plt.xlabel('Arms (t = '+ str(i)+')')
     
    plt.ylabel('The mean')
    plt.xticks(x,column)
    plt.savefig("/Users/salemameen/Desktop/directvsMAB/hedge/Fig"+str(i)+".png")
    plt.clf()
    #plt.show()
```

## Pruning Weights

```python
import math
Threshold = 0
constant = 1
Avg_Accumaltive_R_HGN = np.zeros(16)
Count_HGN = np.zeros(16)
weights = np.ones(16)
import random
E = 0.05
for j in range(17, 1001):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], 
                                                  X_train[b:b+1], 
                                                  batch_size=1, verbose=0)
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], 
                                                          X_train[b:b+1], 
                                                          batch_size=1, 
                                                          verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after 
            Avg_Accumaltive_R_HGN = weights/weights.sum()
            i = np.random.choice(numpy.arange(0, 16), 
                                 p=Avg_Accumaltive_R_HGN.reshape(16))  
            weights[i] = weights[i] * math.pow((1+E), reward) 
            val = Count_HGN[i]
            modelBuckup.set_weights(All_weights)
            Count_HGN[i] = Count_HGN[i] + 1
            ploting_function(j, Avg_Accumaltive_R_HGN, i)
df5_H = pandas.DataFrame(Avg_Accumaltive_R_HGN)
df5_H['Weights']=column
df5_H.set_index('Weights')
df5_H.reset_index('Weights')
df5_H.head(16)
```

```python
import math
gamma = 0.3
Threshold = 0
constant = 1
Avg_Accumaltive_R_EXP3GN = np.zeros(16)
Count_EXP3GN = np.zeros(16)
weights = np.ones(16)
K=16
import random
for j in range(17, 1001):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], 
                                                  X_train[b:b+1], 
                                                  batch_size=1, verbose=0)
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], 
                                                          X_train[b:b+1], 
                                                          batch_size=1, 
                                                          verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after 
            Avg_Accumaltive_R_EXP3GN = (1-gamma)*weights/weights.sum()+gamma/K
            i = np.random.choice(numpy.arange(0, 16), 
                                 p=Avg_Accumaltive_R_EXP3GN.reshape(16))  
            weights[i] = weights[i] * np.exp(gamma*reward/(Avg_Accumaltive_R_EXP3GN[i]*K))
            val = Count_EGN[i]
            modelBuckup.set_weights(All_weights)
            Count_EXP3GN[i] = Count_EXP3GN[i] + 1
            ploting_function(j, Avg_Accumaltive_R_EXP3GN, i)
df5_EXP3 = pandas.DataFrame(Avg_Accumaltive_R_EXP3GN)
df5_EXP3['Weights']=column
df5_EXP3.set_index('Weights')
df5_EXP3.reset_index('Weights')
df5_EXP3.head(16)
```

```python
from scipy.stats import beta
y = np.arange (-0.1, 1.1, 0.001)

def ploting_function(i, S, F):    
    for k in range(16):
        bta = beta.pdf(y, S[k]+1, F[k]+1)
        plt.plot(y, bta)
    plt.xlabel('Arms (t = '+ str(i)+')')

    plt.savefig("/Users/salemameen/Desktop/directvsMAB/Thompson/Fig"+str(i)+".png")
    plt.clf()
```

```python
Avg_Accumaltive_R_TGN = np.zeros(16)
Count_TGN = np.ones(16)
success = np.zeros(16)
failure = np.zeros(16)
import random
ploting_function(0, success, failure)
for i in range(0, 16): 
            loss, accuracy = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[1:10], X_train[1:10], batch_size=1, verbose=0)
            delta = loss - loss_New
            
            
            
            if(delta>0):
                reward = 1
                success[i] = success[i]+1
            else:
                reward = 0
                failure[i] = failure[i]+1                        
            
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            Avg_Accumaltive_R_TGN[i] = reward
            modelBuckup.set_weights(All_weights)
            
            draw = np.random.beta(1+success, 1+failure)
            ploting_function(i+1, success, failure)


for j in range(17, 1001):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
        # Prune the neuron in the layer
            draw = np.random.beta(1+success, 1+failure)
            i = np.argmax(draw)
  
            Count_TGN[i] = Count_TGN[i] + 1
            
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
            delta = loss - loss_New
            
            if(delta>0):
                reward = 1
                success[i] = success[i]+1
            else:
                reward = 0
                failure[i] = failure[i]+1
                
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            
            
            
            val = Count_TGN[i]
            Avg_Accumaltive_R_TGN[i] = (val-1)/val * Avg_Accumaltive_R_TGN[i] + 1/val * reward
            modelBuckup.set_weights(All_weights)
            
            
            ploting_function(j, success, failure)

df5_T = pandas.DataFrame(Avg_Accumaltive_R_TGN)
df5_T['Weights']=column
df5_T.set_index('Weights')
df5_T.reset_index('Weights')
df5_T.head(16)
```

```python
Avg_Accumaltive_R_EXP3GN = np.zeros((16,1))
weights = np.ones((16,1))
Count_EXP3GN = np.ones((16,1))
K=weights.sum()
gama = 0.2
import random

for j in range(0, 1000):
            b = random.randint(0,999) 
            loss, accuracy = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
        # Prune the neuron in the layer
           
            
            prob = (1-gama)*weights/weights.sum() + gama/K
        
        
        
            #prob = np.exp(Avg_Accumaltive_R_EXP3GN/tau)
            #sum = prob.sum()
            #prb = prob/sum
            i = np.random.choice(numpy.arange(0, 16), p=prob.reshape(16))
            Count_EXP3GN[i] = Count_EXP3GN[i] + 1
            
            All_weights=modelBuckup.get_weights()
            temp1 = All_weights[0][0][i]
            temp2 = All_weights[0][1][i]
            temp3 = All_weights[1][i]  # Prune the bias
        # Next Layer
            temp4 = All_weights[2][i][0]
            temp5 = All_weights[2][i][1]
        ######################
            All_weights[0][0][i] = 0
            All_weights[0][1][i] = 0
            All_weights[1][i] = 0  # Prune the bias
        # Next Layer
            All_weights[2][i][0] = 0
            All_weights[2][i][1] = 0
            modelBuckup.set_weights(All_weights)
            loss_New, accuracy_New = modelBuckup.evaluate(X_train[b:b+1], X_train[b:b+1], batch_size=1, verbose=0)
            delta = loss - loss_New
            reward = max(0,Threshold + delta)/constant
            All_weights[0][0][i]= temp1  # before
            All_weights[0][1][i]= temp2  # before
            All_weights[1][i]= temp3   # Bais
            All_weights[2][i][0] = temp4 # after
            All_weights[2][i][1] = temp5   # after
            
            
            weights[i] = weights[i]*np.exp(gama*reward/(prob[i]*K))
            
            modelBuckup.set_weights(All_weights)

df5_EXP3 = pandas.DataFrame(weights)
df5_EXP3['Weights']=column
df5_EXP3.set_index('Weights')
df5_EXP3.reset_index('Weights')
df5_EXP3.head(16)
```
